```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.uniform(low=0, high=1, size=20)
u = np.random.normal(0,0.01,20)
y = 1+2*x+u
plt.figure(figsize=(12, 4)) 
plt.subplot(121)
plt.scatter(x, y)

n = 20
sum_x = np.sum(x)
sum_y = np.sum(y)
x_bar = sum_x/n
y_bar = sum_y/n
beta1 = np.sum((x-x_bar)*(y-y_bar))/np.sum((x-x_bar)*(x-x_bar))
beta0 = y_bar - beta1 * x_bar
plt.subplot(122)
x0 = np.linspace(0, 1, 100)
plt.scatter(x, y)
plt.plot(x0,beta0 + beta1 * x0)
```
