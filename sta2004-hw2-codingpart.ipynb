```python
import numpy as np
import matplotlib.pyplot as plt
y_average = 10
n_list = [1,5,10,30]
fig, ((ax1, ax2, ax3, ax4)) = plt.subplots(4, 1, figsize=(8, 14), sharex=True)
fig_list = [ax1, ax2, ax3, ax4]
color_list = ['blue','green','orange','pink']
for i in range(0,4):
    n = n_list[i]
    ax = fig_list[i]
    colors = color_list[i]
    list_lamta = []
    list_L =[]
    for lamta in np.arange(0.01,1.00,0.01):
        L1 = pow(lamta,n)*np.exp(-lamta*y_average*n)
        lamta_head = 1/y_average
        L2 = pow(lamta_head, n)*np.exp(-lamta_head*y_average*n)
        L = L1 - L2
        list_lamta.append(lamta)
        list_L.append(L)
    ax.set_title('n = {}'.format(n))
    ax.plot(list_lamta, list_L, color=colors )
```

```python
import numpy as np
y_average = 10
lamta_head = 1/y_average
n_list = [1,5,10,30]
for n in n_list:
    left = lamta_head - (lamta_head*1.96)/np.sqrt(n)
    right = lamta_head + (lamta_head*1.96)/np.sqrt(n)
    confidence_interval = np.array([left,right])
    print(confidence_interval)
```

```python
import numpy as np
import matplotlib.pyplot as plt
fig, v = plt.subplots(1, 1, figsize=(12, 9))
v.set_xlabel('lambda')
v.set_ylabel('Samples')
v.vlines(1, 0, 100, colors='g')
for i in range(0,100):
    data = np.random.exponential(1, 10000)
    mean = np.mean(data)
    lamta_head = 1/mean
    left = lamta_head - (lamta_head*1.96)/np.sqrt(10000)
    right = lamta_head + (lamta_head*1.96)/np.sqrt(10000)
    if right<1 or left>1:
        v.hlines(i, left, right, colors='r')
    else:
        v.hlines(i, left, right, colors='k')
plt.show()
```
