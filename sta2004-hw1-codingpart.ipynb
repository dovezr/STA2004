```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.xlim((0, 10))
plt.ylim((0.0, 0.6))
data = np.random.exponential(2.0, 100)
plt.subplot(221)
plt.hist(data, bins=100, density=True)
X1 = np.linspace(0,10,100)
Y1 = 2*(np.exp(-2*X1))
plt.plot(X1,Y1)
mean = np.mean(data) 
var = np.var(data) 
t_data = np.sqrt(100) * (mean - 2) / np.sqrt(var)
print('mean is {}, variance is {},T statistics is {}'.format(mean, var, t_data)) 

t_data_list = []
for i in range (0,1000):
    data1 = np.random.exponential(2.0, 100)
    t_data_list.append(np.sqrt(100) * (np.mean(data1) - 2) / np.sqrt(np.var(data1)))
plt.subplot(222)
plt.hist(t_data_list, bins=100, density=True)
X2 = np.linspace(-4,4,100)
Y2 = np.exp(-X2 ** 2 / 2) / (np.sqrt(2 * np.pi))
plt.plot(X2,Y2)

s_data_list = []
plt.subplot(223)
plt.xlim((0, 1))
plt.ylim((0.0, 2.0))
for i in range (0,1000):
    s_data_list.append(1-norm.cdf(t_data_list[i]))
plt.hist(s_data_list, bins=100, density=True)
X3 = np.linspace(0.00,1.00,100)
Y3 = [1]*100
plt.plot(X3,Y3)
plt.subplot(224)
```

```python
plt.subplot(221)
x = np.random.uniform(low = 0, high = 1.0, size = 100)
eps = np.random.normal(size = 100)
y = np.sin(2 * np.pi * x) + eps
plt.scatter(x, y)

plt.subplot(222)
xs = np.linspace(start = 0, stop = 1.0, num = 20) 
ret1 = np.zeros(20)
ret2 = np.zeros(20)
ret3 = np.zeros(20)
for i in range(20):
    z=xs[i]
    ret1[i] = np.sum((np.absolute(x - z) <= 0.05) * 1.0 * y) / (2.0 * 100 * 0.05)
    ret2[i] = np.sum((np.absolute(x - z) <= 0.15) * 1.0 * y) / (2.0 * 100 * 0.15)
    ret3[i] = np.sum((np.absolute(x - z) <= 0.25) * 1.0 * y) / (2.0 * 100 * 0.25)
plt.plot(xs, ret1, label = 'h = {}'.format(0.05)) 
plt.plot(xs, ret2, label = 'h = {}'.format(0.15)) 
plt.plot(xs, ret3, label = 'h = {}'.format(0.25)) 
plt.legend()

plt.subplot(223)
ret4 = np.zeros(20)
ret5 = np.zeros(20)
ret6 = np.zeros(20)
for i in range(20):
    z=xs[i]
    ret4[i] = np.sum(y/ (100 * 0.05) * (1/np.sqrt(2 * np.pi)) * np.exp(((z - x) / 0.05) ** 2 / -2))
    ret5[i] = np.sum(y / (100 * 0.15) * (1/np.sqrt(2 * np.pi)) * np.exp(((z - x) / 0.15) ** 2 / -2))
    ret6[i] = np.sum(y / (100 * 0.25) * (1/np.sqrt(2 * np.pi)) * np.exp(((z - x) / 0.25) ** 2 / -2))
plt.plot(xs, ret4, label = 'h = {}'.format(0.05)) 
plt.plot(xs, ret5, label = 'h = {}'.format(0.15)) 
plt.plot(xs, ret6, label = 'h = {}'.format(0.25)) 
plt.legend()
plt.subplot(224)   
```
