```python
import numpy as np
import matplotlib.pyplot as plt
u = np.random.normal(0, 1, 200)
X = np.random.uniform(0, 1, (200, 15))
one = np.ones(200)
beta = np.array([1.7,1.2,0,0,0,0.8,0,0,0,-1.5,0,0,0,0,0])
y = one + np.dot(X,beta) + u
plt.scatter(X[:,0], y, label = 'x_1')
plt.scatter(X[:,9], y, label = 'x_10')
plt.legend()
```

```python
import statsmodels.api as sm
X = sm.add_constant(X)
model = sm.OLS(y,X)
results = model.fit()
beta = results.params
y_hat = np.dot(X, beta)
SSE = np.dot((y-y_hat).transpose(),(y-y_hat))
SSR = np.dot((y_hat-np.mean(y)).transpose(),(y_hat-np.mean(y)))
summation = SSE+SSR
TSS = np.dot((y-np.mean(y)).transpose(),(y-np.mean(y)))
R_square = SSR/TSS
errors_variance = SSE/184
print('The SSE is', SSE)
print('The SSR is', SSR)
print('The sum of SSE and SSR is', summation)
print('The TSS is', TSS)
print('The Rˆ2 is', R_square)
print('The errors variance',errors_variance)
```

```python
residual = y-y_hat
plt.grid(color="k", linestyle=":")
plt.scatter(y_hat, residual)
plt.hlines(0,0,4)
```
